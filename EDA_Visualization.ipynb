# Visualization - Analysis
## 1. Univariate Analysis - It will have 1 feature (i.e. 1 column) for analysis.  
###  For example, if you want to see distribution of income, age, speed, etc. 
## 2. Bivariate Analysis - It will have 2 features (i.e. 2 columns) for analysis
###  For example, if you want to analysis on numeric data vs numerical data (Scatter plot is the best option)
###   Numerical data vs categorical data (example: Monthwise revenue (sales))
###   Categorical data vs Categorical data 
## 3. Multivariate Analysis
