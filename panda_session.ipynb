```python
# Create a new column "RideType" based on Miles
# If Miles <5, it's a short ride
# If Miles between 5.1 and 20 then it's a normal ride
# If Miles greater than 20 then it's a long ride
def func(num):
    if num<5:
        return 'Small'
    elif num >= 5 and num < 20:
        return 'Normal'
    else:
        return 'Long' 
    
```

```python
df['Ride Type']=df['MILES*'].apply(func)
```

```python
# Skewness appears when data is spread more on the left hand side or on the right hand side of the mean. 
# Under normal distribution, mean and median will be pretty close to the center 
# When the distribution is skewed on the left side, mean is less than median (i.e. negatively skewed distribution)
# When the distribution is skewed on the right side, mean is higher than median (i.e. positively skewed distribution)
```
